## HeatMap
: It is used to visualize the correlation matrix between two or multiple variables. A heatmap represents the correlation coefficients with colors, allowing for an easy understanding of the relationships between each pair of variables. It can be generated using libraries like Seaborn and Matplotlib.

### RelationShip
 1: Indicates a perfect positive linear correlation, meaning that as one variable increases, the other variable also increases.

0: Indicates no correlation, suggesting that there is no linear relationship between the two variables, although a non-linear relationship may exist.

-1: Indicates a perfect negative linear correlation, meaning that as one variable increases, the other variable decreases.

```python
data=df
var1 = 'Diabetes_binary'
var2 = 'HighBP'
correlation_matrix=df[[var1,var2]].corr()
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix,annot=True, cmap='coolwarm',linewidths=0.5)
plt.title('A heatmap of the correlation matrix between two variables')
plt.show()
# A correlation coefficient of 0.5 or higher indicates a strong correlation.
```

```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report


# Read Data
df = pd.read_csv(r"C:\Users\eunji\OneDrive\Documents\Diabetes.Data.csv")

# Feature and Target Variable Separation
X = df.drop('Diabetes_binary', axis=1)
y = df['Diabetes_binary']

# Split data into train data and test data (70% Training, 30% Test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Data Scaling to appropriately measure the importance
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Logistic Regression Model Creation and Train (class_weight='balanced' to address class unbalance issue)
model = LogisticRegression(max_iter=1000, class_weight='balanced')
model.fit(X_train, y_train)

# Model Prediction
y_pred = model.predict(X_test)

# Model Accuracy Evaluation
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

# Print Results
print(" Accuracy:", accuracy)
print("\n Classification Report:\n", report)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# 2D Data Creation
X, y = make_classification(n_samples=1000, n_features=2, n_informative=1, n_redundant=0, n_repeated=0,n_clusters_per_class=1)

# Data Separation(Training Data 70%, Test Data 30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Data Scailing (StandardScaler Apply)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train Logistic Regression Model (class_weight='balanced'ì™€ max_iter=1000 ì„¤ì •)
model = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
model.fit(X_train, y_train)

# Set boundaries
h = .02  # Grid step size
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot Points
plt.contourf(xx, yy, Z, alpha=0.8)  
plt.scatter(X[y == 0, 0], X[y == 0, 1], c='black', edgecolors='k', marker='o', label="No Diabetes (y=0)")
plt.scatter(X[y == 1, 0], X[y == 1, 1], c='yellow', edgecolors='k', marker='o', label="Diabetes (y=1)")

plt.title("Logistic Regression Decision Boundary (Class Weight Balanced)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier 


df = pd.read_csv(r"C:\Users\eunji\OneDrive\Documents\Diabetes.Data.csv")


X = df.drop('Diabetes_binary', axis=1)  
y = df['Diabetes_binary']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Model Training
model = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=200, random_state=42,early_stopping=True)
model.fit(X_train, y_train)

# Train and Calculate the Model Accuracy
train_accuracy = model.score(X_train, y_train)
test_accuracy = model.score(X_test, y_test)

print("Training Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)

# Predict
y_pred = model.predict(X_test)

# Evaluate Accuracy
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)

# Print Confusion Matrix
confusion = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", confusion)
```
